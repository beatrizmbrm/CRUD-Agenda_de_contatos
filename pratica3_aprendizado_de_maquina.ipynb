```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, train_test_split, ParameterGrid
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
irisdata = pd.read_csv('dados/iris.data', header = None)
```

```python
irisdata
```

```python
irisdata.columns = ['SEPAL_L', 'SEPAL_W', 'PETAL_L', 'PETAL_W', 'CLASSE']
```

```python
irisdata
```

```python
irisdata['CLASSE'] = irisdata['CLASSE'].replace('Iris-setosa', 0)
irisdata['CLASSE'] = irisdata['CLASSE'].replace('Iris-versicolor', 1)
irisdata['CLASSE'] = irisdata['CLASSE'].replace('Iris-virginica', 2)
```

```python
irisdata
```

```python
x = irisdata.drop(['CLASSE'], axis = 1)
x = x.values
y = irisdata['CLASSE']
```

```python
kf = KFold(n_splits=5, shuffle = True, random_state = 42)
kf.get_n_splits(x)
```

```python
param_grid = {'n_neighbors': [3, 5, 7, 9, 11, 15], 'metric' : ['euclidean', 'manhattan', 'minkowski', 'cosine']}

test_scores = []
```

```python
print(kf)
for i, (train_index, test_index) in enumerate(kf.split(x)):

    x_train_full, x_test = x[train_index], x[test_index]
    y_train_full, y_test = y.iloc[train_index], y.iloc[test_index]

    x_train , x_val, y_train, y_val = train_test_split(
        x_train_full, y_train_full, test_size=0.2, random_state=42)

    accs_val = []#acuracia
    par = []#hiperparametros

    print(f"Fold {i}:")
    print(f"  Train: index={train_index}")
    print(f"  Test:  index={test_index}")

    for params in ParameterGrid(param_grid):
        knn = KNeighborsClassifier(metric = params['metric'], n_neighbors = params['n_neighbors'])
        knn.fit(x_train, y_train)
        y_pred = knn.predict(x_val)
        acc = accuracy_score(y_val, y_pred)
        accs_val.append(acc)
        par.append(params)


    print(par[accs_val.index(max(accs_val))])

    knn = KNeighborsClassifier(metric = par[accs_val.index(max(accs_val))]['metric'], n_neighbors = par[accs_val.index(max(accs_val))]['n_neighbors'])
    knn.fit(x_train_full, y_train_full)
    y_pred = knn.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    test_scores.append(acc)
    print(acc)


```

```python
print(f"A acuracia media nos conjuntos de testes: {sum(test_scores) / len(test_scores):.4f}")
```
